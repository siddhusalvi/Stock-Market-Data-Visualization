```scala
import vegas._
import vegas.render.WindowRenderer._
import vegas.sparkExt._
import vegas.data.External._
```

```scala
def sc = spark.sparkContext
```

```scala
val stocksDF = spark.read
  .option("inferSchema","true")
  .option("header","true")
  .csv("stockdata.csv")
```

```scala
Vegas("Example of Line chart")
      .withDataFrame(stocksDF)
      .mark(Line)
      .encodeX("date", Temp)
      .encodeY("high", Quantitative)
      .configCell(width = 300.0, height = 200.0)
      .show
```

```scala
Vegas("Example of Bar chart")
      .withDataFrame(stocksDF)
      .encodeX("date", Temp)
      .encodeY("volume", Quantitative)
      .mark(Bar)
      .show
```

```scala
Vegas("Example of Area chart")
      .withDataFrame(stocksDF)
      .mark(Area)
      .encodeX("date", Temp)
      .encodeY("volume", Quantitative)
      .configCell(width = 300.0, height = 200.0)
      .show
```
